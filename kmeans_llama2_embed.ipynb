```python
import pickle

llama2_embed_file = '/tmp2/cychang/llama2_embedding.pkl'
with open(llama2_embed_file, 'rb') as fp:
    llama2_embedding = pickle.load(fp)

llama2_embedding
```

```python
llama2_tensor = [v for k, v in llama2_embedding.items()]
llama2_tensor
```

```python
from sklearn.cluster import MiniBatchKMeans

kmeans = MiniBatchKMeans(n_clusters=1000, batch_size=10240)
result = kmeans.fit(llama2_tensor)
result.labels_
```

```python
from collections import Counter
Counter(list(result.labels_)).most_common()
```
